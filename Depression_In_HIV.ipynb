# Step 3 Categorical Variables

```python
categorical_vars = [
"2.Sex",
"4.Marital Status",
"7.Occupation",
"8.Employment Status",
"6.Level of education",
"21. Perception",
"22.Knowledge",
"20. Loss of trust",
"19.Confidentiality",
"17.Stigma",
"18. Discrimination",
"current_regimen",
"cd4_category"
]
```

```python
for var in categorical_vars:
    print(var)
    print(df[var].value_counts(dropna=False))
    print("-"*30)
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

results = []

for var in categorical_vars:
    # Ensure the column is string and strip spaces
    df[var] = df[var].astype(str).str.strip()

    # Skip if only one unique value or all NaN
    if df[var].nunique(dropna=True) <= 1:
        print(f"Skipping {var} — only one unique value")
        continue

    # Generate crosstab
    table = pd.crosstab(df[var], df["depression"])

    # Skip if table is empty
    if table.size == 0:
        print(f"Skipping {var} — crosstab is empty")
        continue

    # Run chi-square test
    chi2, p, dof, expected = chi2_contingency(table)

    results.append({"Variable": var, "Chi2": chi2, "p_value": p})

# Create DataFrame and sort by p-value
results_df = pd.DataFrame(results).sort_values("p_value")
print(results_df)
```

```python
# Variables with p-value <0.20 are significamt and will be used in the multivariate analysis
significant_vars = results_df[results_df["p_value"] < 0.20]

print(significant_vars)
```

```python
results_df.to_excel("bivariate_results.xlsx", index=False)
```

If p < 0.05, there is a significant association. Include variables with p < 0.20 in multivariate model

# Step 4 Continuous Variables


```python
continuous_vars = [
"1.Age in years",
"cd4_count",
"Valid Viral Load (VL)",
"duration_years",
                    "total_ART_years"
]

# T-test analysis
from scipy.stats import ttest_ind

results_cont = []

for var in continuous_vars:

    group1 = df[df["depression"] == 1][var]
    group0 = df[df["depression"] == 0][var]

    stat, p = ttest_ind(group1, group0, nan_policy="omit")

    results_cont.append({
        "Variable": var,
        "t_stat": stat,
        "p_value": p
    })

results_cont_df = pd.DataFrame(results_cont)

results_cont_df.sort_values("p_value")
```

# Step 4 Extract Odd Ratios and CI and P-values

```python
import numpy as np

params = result.params
conf = result.conf_int()

OR = np.exp(params)
CI_lower = np.exp(conf[0])
CI_upper = np.exp(conf[1])

summary_table = pd.DataFrame({
    "Odds Ratio": OR,
    "CI Lower": CI_lower,
    "CI Upper": CI_upper,
    "p-value": result.pvalues
})

print(summary_table)
```
